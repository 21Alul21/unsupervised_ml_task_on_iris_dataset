```python
# using elbow method to confirm the perfect number of clusters, although the dataset indicates 3
k_range = range(1, 6)
SSE = []

for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(x_scaled)
    SSE.append(km.inertia_)
    
```

```python
SSE
```

```python
plt.plot(np.arange(1,6), SSE)
plt.xlabel('K(number of clusters)')
plt.ylabel('SSE')
plt.title('Elbow method for choosing best K number')
plt.show()
```

```python
# since the dataset indicates three clusters, we will go with 3 clusters: setosa, versicolor and virginica
km = KMeans(n_clusters=3, random_state=42)
clusters = km.fit_predict(x_scaled)
```
